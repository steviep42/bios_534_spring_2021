## Installing Python

Once upon a time, this used to be simple and straightforward because for the longest time, Python was used almost exclusively on UNIX / LINUX systems wherein someone installed the language for you. Of course, you would then need to install any additional "modules" (we'll get to those soon) that you might need. However, for the most part you didn't really do much to get access to the language except access the server upon which it was installed. You relied mostly upon the server administrators to manage the system and any development languages that might become necessary.

### Version conflicts

Over time, people started building module add ons that could make life simple for you by replacing some key aspect of your work flow. The same was/us true of R. The problem with this is that a cool new addon might require you to install a new or updated version of Python just to be able to run the new addon. So on a server you would then bug the administrator to 1) either upgrade the version of Python or 2) leave the existing version of Python intact and 2) install the new version along side the existing one though make it accessible by making reference to another command name (e.g. python3.7 vs python3.6). This is a headache in any language. 

### Python and Laptops

As Python continued to rise in popularity, it became available for Apple Mac OSX as part of the internal OS management tools and eventually onto MS Windows. Of course, if its on **your** laptop then you become the administrator which means that **you** have to manage the problems just described. On OSX there are tools such a [**Homebrew**](https://brew.sh/) that can help you manage Python and other tools traditionally associcated with UNIX / LINUX but that still doesn't help you so much if you need to manage multiple versions of Python. 

### Virtual Environments

One solution that was introduced to help deal with this issue was "virtual environments". Virtual environments are commonly used to run different versions of programs or to test a given version of a program with different versions of its dependencies. This includes different versions of Python itself.
Don't get too confused here but this is a very real problem experienced by Python developers the world over. People who use R can also have the same problem. You find a great package on CRAN that solves most of your problems only to find out that you need to upgrade to the latest version of R to get it to run ! Most developers, however, if they are being reasonable, will develop packages such that they run with the most recent version of the language as well as one or two versions back. But you cannot count on this.

## Walk Through Using Some Engine Emission Data

Let's look at some data that lists engines and the emmission output coming from them. Note also how easy it is to refer to files on the Internet and read them in directly. This is particularly true with .csv files which can generally be read by any number of programs and languages such as Excel, R, Python, Java, etc. The advantage is that there is no need to first download the file and then read it, unless you want to do that for some reason. 

````python
# The pandas package gives "data frame" capability to Python
import pandas as pd
url = "https://raw.githubusercontent.com/steviep42/youtube/master/YOUTUBE.DIR/table_7_3.csv"

# Read in the CSV file form the URL
engine = pd.read_csv(url)

# Check out the first few rows
engine.head()

# Get some basic summary statistics for each column
engine.describe()
````

## Checking Distributions

So one thing we could do is try to get a handle on the range and distribution of the various gasses emanating from and engine. Using boxplots is a popular choice. Here we see that Carbon Monoxide exhibits a wide range of values in a way that the other two do not. This does not mean that the other two, nox and hc, are normally distibuted but it does suggest perhaps that "co" is not. 

````python
# Drop the "en" column since it is simply a number which is assigned to each engine
engine_noeng = engine.drop(['en'],axis=1)

# Check the first few rows
engine_noeng.head()

# Pandas data frames have a built-in boxplot method
engine_noeng.boxplot()
````

## Plotting The Data

So from here we could create a qq-plot, boxplot, and histogram of the "co" (carbon monoxide) column to try to make some guess as to whether the data are normally distributed. We'll use the **matplotlib** graphics module to do this. 

````python
from matplotlib import pyplot as plt
import scipy.stats as stats

# Create a plot with 3 sub plots
fig, (ax1,ax2,ax3) = plt.subplots(1,3,tight_layout=True)

# Create a QQ Plot
stats.probplot(engine_noeng.co,plot=ax1)

# Create a boxplot
ax2.boxplot(engine_noeng.co)

# Set a title
ax2.set_title('co boxplot')

# Create a histogram
ax3.hist(engine_noeng.co)

# Set a title
ax3.set_title('co histogram')
````

## Check for Normality of Data

At first glance, this does not appear to be normally distributed. To help out, we could also use some standard tests. One such test is called the [Shapiro-Wilk](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html) test which tests the null hypothesis that the data was drawn from a normal distribution. So, the default assumption is that it is - unless we are presented with evidence, in this case a "low" p-value, that suggests otherwise.

````python
from scipy import stats
stats.shapiro(engine_noeng.co)
````

## Transformations

One approach with data is to apply some transformation to it that might result in the transformed data being normally distributed. One possible transformtion is the logarithm function. It's easy to apply and then use the same Shapiro-Wilk test as before. 


````python
import numpy as np
stats.shapiro(np.log(engine_noeng.co))
````

So here it looks like we would hold on to the Null Hypothesis that the data is normally distributed. We could go ahead and replot the data to see how the log-transformed data looks visually. 

````python
logeng = np.log(engine_noeng.co)
fig, (ax1,ax2,ax3) = plt.subplots(1,3,tight_layout=True,figsize=(8,5))
stats.probplot(logeng,plot=ax1)
ax2.boxplot(logeng)
ax2.set_title('co boxplot')
ax3.hist(logeng)
ax3.set_title('co histogram')
````

We might even create a confidence interval for the mean of the log transformed carbon monoxide data.
$$ \bar{x}  \pm t_{n-1}\frac{s}{\sqrt{n}} $$

````python
import statistics
logco = np.log(engine_noeng.co)

# Get the man and standard deviation for the data
my_mean = statistics.mean(logco)
my_sd = statistics.stdev(logco)

# Get the lenght of the data
n = engine_noeng.shape[0]

# Compute the standard error
se = my_sd / np.sqrt(n)

# Get the associated PDF value
from scipy.stats import t
error = se * t.ppf(0.975,n-1)

# Create the interval
left  = my_mean - error
right = my_mean + error

print('Log Transformed: left=%.3f, right=%.3f' % (left, right))

# For the non log transformed data this would be
print('Non Log Transformed: left=%.3f, right=%.3f' % (np.exp(left), np.exp(right)))

````

## Hypothesis Test

So let's create a little test here. Basically let's test the hypothesis that a valuer of 5.4 represents the true mean of the carbon monoxide data. To this, we'll need to log transform it of course. 

````python
lNull = np.log(5.4) - error
rNull = np.log(5.4) + error

print(round(lNull,2),round(rNull,2))
print(my_mean)
````

So the mean is outside the range thus we reject the null. There is a low probability that we would have obtained our sample mean if the true mean really was 5.4. We could have also calculated a p-value by hand. 

````python
p_val = 2*(1-t.cdf((my_mean-np.log(5.4))/se,df=n-1))
print(p_val)
````

## Use the Built in Test

However, we don't need to do all that work by hand when we can use a built in T-Test to do this for us. One of the differences between Python and R is that R has lots of statistical tests and distribution functions loaded by default which some claim make it easier to "do stats" on the fly. To an extent this is true because there are a number of Python modules that supply statistical testing functions. You have to know what they are called and then load them.

````python
tStat, pValue =  stats.ttest_1samp(np.log(engine_noeng.co), np.log(5.4), axis=0)
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) #print the P-Value and the T-Statistic
###P-Value:0.5340949682112062 T-Statistic:0.6292755379958038
````
